```python
#Select behavioral features
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

clustering_features = ['Time_Spent_Total_Minutes', 'Saved_Event_Count', 'Feedback_Rating', 'Days_Since_Last_Activity']
clustering_data = df[clustering_features].copy()
clustering_data.fillna(0, inplace=True)

#Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

#Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['User_Segment'] = kmeans.fit_predict(scaled_data)

#Visualize clustering results
plt.figure(figsize=(8,6))
scatter = plt.scatter(df['Time_Spent_Total_Minutes'], df['Saved_Event_Count'],
                      c=df['User_Segment'], cmap='viridis', alpha=0.6, s=40)
plt.xlabel('Time Spent on Platform')
plt.ylabel('Saved Event Count')
plt.title('User Segmentation via KMeans Clustering')
plt.grid(True)
plt.legend(*scatter.legend_elements(), title="Segments")
plt.show()
```

```python
# Group by Platform Source and calculate key engagement metrics
platform_analysis = df.groupby('Platform_Source')[['Course_Completed', 'Time_Spent_Total_Minutes', 'Feedback_Rating', 'Days_Since_Last_Activity']].agg({
    'Course_Completed': 'mean',
    'Time_Spent_Total_Minutes': 'mean',
    'Feedback_Rating': 'mean',
    'Days_Since_Last_Activity': 'mean'
}).round(2)

# Rename for clarity
platform_analysis.rename(columns={
    'Course_Completed': 'Avg_Completion_Rate',
    'Time_Spent_Total_Minutes': 'Avg_Time_Spent',
    'Feedback_Rating': 'Avg_Feedback',
    'Days_Since_Last_Activity': 'Avg_Days_Inactive'
}, inplace=True)

# Sort by completion rate (or any metric you want to highlight)
platform_analysis.sort_values(by='Avg_Completion_Rate', ascending=False)
```

```python
import matplotlib.pyplot as plt

# Bar chart for Avg Completion Rate by Platform Source
plt.figure(figsize=(10,6))
df_roi = platform_analysis.sort_values(by='Avg_Completion_Rate', ascending=False)

plt.bar(df_roi.index, df_roi['Avg_Completion_Rate'], color='skyblue')
plt.title('Average Course Completion Rate by Marketing Source', fontsize=14)
plt.xlabel('Platform Source', fontsize=12)
plt.ylabel('Avg Completion Rate', fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 0.5)  # adjusting based on the data range
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
